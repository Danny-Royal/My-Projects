#Data Prediction using **dot notation**

```python
melb_data = drop_missing_values
melb_data
```

```python
y = melb_data.Price
```

```python
melb_data.columns
```

```python
melb_features = ["Price", "Cost", "Revenue"]
```

```python
X = melb_data[melb_features]
```

```python
X.describe()
```

```python
X.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
melb_model = DecisionTreeRegressor(random_state=1)
melb_model.fit(X,y)
```

```python
melb_model.predict(X.head())
```

```python
print("Making predictions for the following 5 Item types:")
print(X.head())
print("The predictions are")
print(melb_model.predict(X.head()))
```
